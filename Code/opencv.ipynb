```python
import cv2
import os
import math

# Function to calculate the greatest common divisor (GCD)
def gcd(a, b):
    return math.gcd(a, b)

# Directory to input images
input_folder = "C:/Users/balav/Downloads/Solution_2/Solution_2/Test/Gravity_AI/Logo"

# Loop through each image in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith('.png'):
        img_path = os.path.join(input_folder, filename)
        
        # Read the image
        img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
        
        # Get the resolution (height, width)
        height, width = img.shape[:2]
        
        # Calculate the aspect ratio
        ratio_gcd = gcd(width, height)
        aspect_ratio = f"{width // ratio_gcd}:{height // ratio_gcd}"
        
        # Print the resolution and aspect ratio in the required format
        if height == width:
            print(f"The image {filename} is a square ({width}x{height}). Convert it to 100x100 resolution.")
        else:
            print(f"The image {filename} is a rectangle ({width}x{height}). Aspect Ratio: {aspect_ratio}. Convert it to 256x77 resolution.")
```
